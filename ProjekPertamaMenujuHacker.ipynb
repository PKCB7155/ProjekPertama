```python
from IPython.display import display, HTML
import time
import random

# HTML + JavaScript untuk Matrix Rain (sangat smooth di Colab)
html_code = """
<!DOCTYPE html>
<html>
<head>
    <style>
        body {
            background-color: black;
            margin: 0;
            overflow: hidden;
        }
        canvas {
            display: block;
        }
    </style>
</head>
<body>
    <canvas id="matrix"></canvas>

    <script>
        const canvas = document.getElementById("matrix");
        const ctx = canvas.getContext("2d");

        // Saiz canvas ikut Colab
        function resizeCanvas() {
            canvas.height = window.innerHeight * 0.85;
            canvas.width = window.innerWidth * 0.95;
        }
        resizeCanvas();
        window.addEventListener("resize", resizeCanvas);

        const chars = "01アイウエオカキクケコサシスセソタチツテトナニヌネノハヒフヘホマミムメモヤユヨラリルレロワン0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ";
        const fontSize = 14;
        const columns = canvas.width / fontSize;
        const drops = Array(Math.floor(columns)).fill(1);

        function draw() {
            ctx.fillStyle = "rgba(0, 0, 0, 0.05)";
            ctx.fillRect(0, 0, canvas.width, canvas.height);

            ctx.fillStyle = "#00ff41";
            ctx.font = fontSize + "px monospace";

            for (let i = 0; i < drops.length; i++) {
                const text = chars[Math.floor(Math.random() * chars.length)];
                ctx.fillText(text, i * fontSize, drops[i] * fontSize);

                if (drops[i] * fontSize > canvas.height && Math.random() > 0.975) {
                    drops[i] = 0;
                }
                drops[i]++;
            }
        }

        setInterval(draw, 35);
    </script>
</body>
</html>
"""

display(HTML(html_code))
```

Flappy Bird


```python
from IPython.display import display, HTML
import time

html_flappy = """
<!DOCTYPE html>
<html>
<head>
    <style>
        body { background: #000; margin: 0; overflow: hidden; font-family: Arial; }
        canvas { display: block; margin: 20px auto; background: #70c5ee; }
        #score { position: absolute; top: 20px; left: 50%; transform: translateX(-50%); color: white; font-size: 50px; font-weight: bold; text-shadow: 3px 3px 0 #000; }
        #start { position: absolute; top: 50%; left: 50%; transform: translate(-50%, -50%); color: white; font-size: 30px; text-align: center; cursor: pointer; }
    </style>
</head>
<body>
    <div id="score">0</div>
    <canvas id="game" width="400" height="600"></canvas>
    <div id="start">Tekan SPACE atau Klik untuk Terbang<br><small>(Klik di sini untuk mula)</small></div>

    <script>
        const canvas = document.getElementById("game");
        const ctx = canvas.getContext("2d");
        const scoreEl = document.getElementById("score");
        const startEl = document.getElementById("start");

        let birdY = 250, birdVel = 0, gravity = 0.5, jump = -12;
        let pipes = [], frame = 0, score = 0, gameOver = false, started = false;

        function randomPipe() {
            const gap = 160;
            const height = Math.random() * 200 + 100;
            return { x: 400, top: height, bottom: height + gap };
        }

        pipes.push(randomPipe());

        function draw() {
            // Langit
            ctx.fillStyle = "#70c5ee";
            ctx.fillRect(0, 0, 400, 600);

            // Tanah
            ctx.fillStyle = "#5e9c3d";
            ctx.fillRect(0, 520, 400, 80);

            // Burung
            ctx.fillStyle = "#ffeb3b";
            ctx.beginPath();
            ctx.arc(100, birdY, 18, 0, Math.PI * 2);
            ctx.fill();
            ctx.fillStyle = "#000";
            ctx.beginPath();
            ctx.arc(110, birdY - 5, 6, 0, Math.PI * 2);
            ctx.fill();

            // Pipes
            ctx.fillStyle = "#4caf50";
            for (let pipe of pipes) {
                ctx.fillRect(pipe.x, 0, 60, pipe.top);
                ctx.fillRect(pipe.x, pipe.bottom, 60, 600);
            }

            // Score
            scoreEl.textContent = score;
        }

        function update() {
            if (!started || gameOver) return;

            birdVel += gravity;
            birdY += birdVel;

            // Pipe movement
            for (let pipe of pipes) {
                pipe.x -= 2;

                // Score
                if (pipe.x + 60 < 100 && !pipe.scored) {
                    score++;
                    pipe.scored = true;
                }

                // Collision
                if (100 + 18 > pipe.x && 100 < pipe.x + 60) {
                    if (birdY - 18 < pipe.top || birdY + 18 > pipe.bottom) {
                        gameOver = true;
                    }
                }
            }

            // Tambah pipe baru
            if (frame % 90 === 0) pipes.push(randomPipe());

            // Buang pipe lama
            if (pipes[0].x < -60) pipes.shift();

            // Collision tanah & langit
            if (birdY > 500 || birdY < 0) gameOver = true;

            frame++;
        }

        function loop() {
            update();
            draw();
            requestAnimationFrame(loop);
        }

        // Kawalan
        function flap() {
            if (!started) {
                started = true;
                startEl.style.display = "none";
            }
            if (!gameOver) birdVel = jump;
            else location.reload(); // Restart
        }

        document.addEventListener("keydown", e => { if (e.code === "Space") flap(); });
        canvas.addEventListener("click", flap);
        startEl.addEventListener("click", flap);

        loop();
    </script>
</body>
</html>
"""

display(HTML(html_flappy))
```

SNAKE GAME

```python
from IPython.display import display, HTML

html_snake = """
<!DOCTYPE html>
<html>
<head>
    <style>
        body {
            background: #111;
            color: white;
            font-family: 'Courier New', monospace;
            text-align: center;
            margin: 0;
        }
        canvas {
            border: 4px solid #0f0;
            box-shadow: 0 0 20px #0f0;
            margin: 20px auto;
            background: #000;
        }
        #score {
            font-size: 28px;
            margin: 10px;
        }
        #gameover {
            position: absolute;
            top: 50%;
            left: 50%;
            transform: translate(-50%, -50%);
            font-size: 50px;
            color: #f00;
            display: none;
            text-shadow: 0 0 10px #f00;
        }
    </style>
</head>
<body>
    <h1>🐍 SNAKE GAME</h1>
    <div id="score">Score: 0</div>
    <canvas id="canvas" width="600" height="600"></canvas>
    <div id="gameover">GAME OVER<br><small>Tekan SPACE untuk main semula</small></div>

    <script>
        const canvas = document.getElementById("canvas");
        const ctx = canvas.getContext("2d");
        const scoreEl = document.getElementById("score");
        const gameOverEl = document.getElementById("gameover");

        const grid = 20;
        let snake = [{x: 300, y: 300}];
        let dx = grid, dy = 0;
        let food = {x: 400, y: 400};
        let score = 0;
        let gameOver = false;
        let gameInterval;

        function randomFood() {
            food.x = Math.floor(Math.random() * (canvas.width / grid)) * grid;
            food.y = Math.floor(Math.random() * (canvas.height / grid)) * grid;
        }

        function draw() {
            // Background
            ctx.fillStyle = "#000";
            ctx.fillRect(0, 0, canvas.width, canvas.height);

            // Snake
            snake.forEach((segment, i) => {
                ctx.fillStyle = i === 0 ? "#0f0" : "#0a0";
                ctx.fillRect(segment.x, segment.y, grid, grid);
                ctx.strokeStyle = "#000";
                ctx.strokeRect(segment.x, segment.y, grid, grid);
            });

            // Food
            ctx.fillStyle = "#f00";
            ctx.beginPath();
            ctx.arc(food.x + grid/2, food.y + grid/2, grid/2, 0, Math.PI * 2);
            ctx.fill();
        }

        function update() {
            if (gameOver) return;

            const head = {x: snake[0].x + dx, y: snake[0].y + dy};

            // Collision wall
            if (head.x < 0 || head.x >= canvas.width || head.y < 0 || head.y >= canvas.height) {
                endGame();
                return;
            }

            // Collision self
            for (let segment of snake) {
                if (head.x === segment.x && head.y === segment.y) {
                    endGame();
                    return;
                }
            }

            snake.unshift(head);

            // Eat food
            if (head.x === food.x && head.y === food.y) {
                score += 10;
                scoreEl.textContent = `Score: ${score}`;
                randomFood();
            } else {
                snake.pop();
            }

            draw();
        }

        function endGame() {
            gameOver = true;
            clearInterval(gameInterval);
            gameOverEl.style.display = "block";
        }

        function startGame() {
            snake = [{x: 300, y: 300}];
            dx = grid;
            dy = 0;
            score = 0;
            scoreEl.textContent = "Score: 0";
            gameOver = false;
            gameOverEl.style.display = "none";
            randomFood();
            clearInterval(gameInterval);
            gameInterval = setInterval(update, 80);
            draw();
        }

        // Controls
        document.addEventListener("keydown", e => {
            if (gameOver && e.key === " ") {
                startGame();
                return;
            }

            switch(e.key) {
                case "ArrowUp":    if (dy !== grid) { dx = 0; dy = -grid; } break;
                case "ArrowDown":  if (dy !== -grid) { dx = 0; dy = grid; } break;
                case "ArrowLeft":  if (dx !== grid) { dx = -grid; dy = 0; } break;
                case "ArrowRight": if (dx !== -grid) { dx = grid; dy = 0; } break;
            }
        });

        // Start pertama kali
        startGame();
    </script>
</body>
</html>
"""

display(HTML(html_snake))
```
